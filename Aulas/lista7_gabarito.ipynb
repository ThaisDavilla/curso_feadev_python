**curiosidade: oq siginifica autopct='%1.1f%%**
    
O parmetro autopct em um grfico de pizza (pie chart) do Matplotlib  usado para formatar a exibio das porcentagens das fatias no grfico. Ele aceita uma string de formatao que controla como as porcentagens so exibidas.

%1.1f%%: Vamos dividi-lo em partes:
%1.1f: Este  um especificador de formato que indica que queremos exibir um nmero de ponto flutuante (float) com uma casa decimal.
%%: Este  usado para exibir o smbolo de porcentagem "%" literalmente.
Portanto, %1.1f%% significa que estamos formatando o nmero de ponto flutuante com uma casa decimal e, em seguida, adicionamos o smbolo de porcentagem "%" no final.

Por exemplo, se a porcentagem da fatia for 25.4%, autopct='%1.1f%%' far com que ela seja exibida como "25.4%". Isso ajuda a fornecer uma exibio mais precisa das porcentagens no grfico de pizza. Voc pode ajustar a formatao conforme necessrio para atender s suas preferncias.

### Qual  a proporo de sobreviventes por sexo?

Passo a Passo:

**Calcule a contagem de sobreviventes por sexo. (utilize groupby)**

Neste passo, voc est agrupando os dados por sexo (gnero) e, em seguida, calculando a contagem de sobreviventes para cada grupo. Isso significa que voc est contando quantos passageiros sobreviveram para cada categoria de sexo (masculino e feminino).


**Calcule a proporo de sobreviventes para cada sexo**

Agora, voc est calculando a proporo de sobreviventes para cada categoria de sexo. Isso  feito dividindo o nmero de sobreviventes para cada sexo pelo total de sobreviventes.

**Exiba a proporo de sobreviventes por sexo**

Neste passo, voc est simplesmente exibindo a proporo de sobreviventes por sexo. Isso mostrar a proporo de passageiros sobreviventes do sexo masculino e feminino em relao ao total de sobreviventes.


**Crie um grfico de barras para representar a proporo de sobreviventes por sexo**


```python
### seu cdigo comea aqui##

import matplotlib.pyplot as plt

# Calcule a contagem de sobreviventes por sexo (passo 1)
contagem_sobreviventes_por_sexo = titanic_data.groupby('Sexo')['Sobreviveu'].sum()

# Calcule a proporo de sobreviventes para cada sexo (passo 2)
total_sobreviventes = titanic_data['Sobreviveu'].sum()
proporcao_sobreviventes_por_sexo = contagem_sobreviventes_por_sexo / total_sobreviventes

# Exiba a proporo de sobreviventes por sexo (passo 3)
print("Proporo de Sobreviventes por Sexo:")
print(proporcao_sobreviventes_por_sexo)

# Crie um grfico de barras para representar a proporo de sobreviventes por sexo (passo 4)
proporcao_sobreviventes_por_sexo.plot(kind='bar', color=['blue', 'pink'])
plt.title('Proporo de Sobreviventes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Proporo de Sobreviventes')
plt.show()
```

###  Qual  a mdia de idade dos passageiros que sobreviveram em comparao com os que no sobreviveram? 

Passo a Passo:

**Calcule a mdia de idade dos passageiros que sobreviveram e dos que no sobreviveram**

Primeiro, calcule a mdia de idade dos passageiros que sobreviveram e dos que no sobreviveram. 

**Crie um grfico de barras para visualizar a diferena na mdia de idade**

Agora, crie um grfico de barras para visualizar a diferena na mdia de idade entre os grupos de sobreviventes e no sobreviventes

```python
### seu cdigo comea aqui##


# Calcule a mdia de idade dos passageiros que sobreviveram e dos que no sobreviveram (passo 1)
media_idade_sobreviventes = titanic_data[titanic_data['Sobreviveu'] == 1]['Idade'].mean()
media_idade_nao_sobreviventes = titanic_data[titanic_data['Sobreviveu'] == 0]['Idade'].mean()

# Exiba a mdia de idade dos passageiros que sobreviveram e dos que no sobreviveram
print("Mdia de Idade dos Sobreviventes:", media_idade_sobreviventes)
print("Mdia de Idade dos No Sobreviventes:", media_idade_nao_sobreviventes)

# Crie um grfico de barras para visualizar a diferena na mdia de idade (passo 2)
media_idades = [media_idade_sobreviventes, media_idade_nao_sobreviventes]
grupos = ['Sobreviventes', 'No Sobreviventes']

plt.bar(grupos, media_idades, color=['green', 'red'])
plt.title('Mdia de Idade dos Sobreviventes e No Sobreviventes')
plt.xlabel('Grupo')
plt.ylabel('Mdia de Idade')
plt.show()
```

### Qual  a quantidade de sobreviventes em cada classe social?

**Calcule o nmero de sobreviventes em cada classe social**

Primeiro, voc precisa agrupar os dados pelo atributo "Pclass" (classe social) e contar o nmero de sobreviventes em cada classe. Voc pode fazer isso da seguinte maneira

**Crie um grfico de pizza para representar a quantidade de sobreviventes em cada classe social**

Agora, voc pode criar um grfico de pizza para visualizar a distribuio dos sobreviventes em cada classe social

```python
### seu cdigo comea aqui##

# Calcule o nmero de sobreviventes em cada classe social (passo 1)
sobreviventes_por_classe = titanic_data[titanic_data['Sobreviveu'] == 1].groupby('Classe').size()

# Exiba o nmero de sobreviventes em cada classe social
print("Nmero de Sobreviventes por Classe Social:")
print(sobreviventes_por_classe)

# Crie um grfico de pizza para representar a quantidade de sobreviventes em cada classe social (passo 2)
plt.pie(sobreviventes_por_classe, labels=sobreviventes_por_classe.index, autopct='%1.1f%%', startangle=90)
plt.title('Quantidade de Sobreviventes por Classe Social')
plt.show()
```
